```python
import random
import numpy as np
import matplotlib.pyplot as plt

def epsilon_greedy(k, epsilon, Q, n):
    if random.random() < epsilon:
        # Explore - select a random arm
        return random.randint(0, k-1)
    else:
        # Exploit - select the arm with the highest expected reward
        return np.argmax(Q)

'''

# k = number of arms
# epsilon = exploration probability
# Q = expected reward for each arm
# n = number of steps

'''

k = 10
epsilon = 0.1
Q = [0.0 for i in range(k)]
n = 1000
N = [0 for i in range(k)]
rewards = []

for t in range(n):
    a = epsilon_greedy(k, epsilon, Q, n)
    r = random.gauss(mu=Q[a], sigma=1.0)
    N[a] = N[a] + 1
    Q[a] = Q[a] + (r - Q[a]) / N[a]
    rewards.append(r)
print(rewards)

plt.plot(rewards)
plt.xlabel('Steps')
plt.ylabel('Rewards')
plt.title('Rewards received from epsilon-greedy algorithm')
plt.show()
```
